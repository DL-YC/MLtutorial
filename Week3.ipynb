# Classification

### Using linear regression

$$h_\theta (x) = \theta^T x$$

### Using sigmoid function (AKA, logistic function)

$$g(z) = \frac {1} {1 + e^{-z}}$$

$$ h_\theta (x) = \frac {1} {1 + e^{-\theta^T x}}$$




```python
# This is just a plot of the sigmoid function

def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a

x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
plt.plot(x,sig)
plt.show()
```

The value of $h_\theta(x)$ is interpreted as the probability that y = 1, for input x.

$$ h_\theta(x) = P (y = 1 \ | \ x;  \theta) $$

Accordingly:

$$ P (y = 0 \ | \ x;  \theta) = 1 - h_\theta(x) $$

## Decision Boundary

- Just a threshold
- Predict $y = 1$ is $h_\theta(x) \geq 0.5$
    - (this happens whenever $\theta^T x \geq 0$



```python
### Data for a decision boundary example
import matplotlib.cm as cm

n = 7
x1 = 2 * np.random.rand(1,n)
x2 = 2 * np.random.rand(1,n) + 1.5
xs_linear = np.array([x1, x2])

y1 = 2.5 * np.random.rand(1,n)
y2 = 2.5 * np.random.rand(1,n) + 1
ys_linear = np.array([y1, y2])

colors = cm.rainbow(np.linspace(0, 1, len(ys_linear)))
for x, y, c in zip(xs_linear, ys_linear, colors):
    plt.scatter(x, y, color=c)
plt.show()
```

To create the decision boundary for the above example, use logistic regression:

$$h_\theta(x) = g(\theta_0 + \theta_1 x_1 + \theta_2 x_2)$$

We will guess that our parameters are: 

$$
\theta = 
\begin{bmatrix}
-3\cr
1\cr
1
\end{bmatrix}$$

Predict $y=1$ if 

$$\theta^T x \geq 0$$
$$-3 + x_1 + x_2 \geq 0$$

Where our decision boundary is

$$x_1 + x_2 = 3$$

```python
### Plotting decision boundary on the data
linex = np.arange(0, 3.5, 0.2)
theta = np.array([-3, 1, 1])
liney = (- theta[1] * linex - theta[0] * 1) / theta[2]
plt.plot(linex, liney, color='k')
colors = cm.rainbow(np.linspace(0, 1, len(ys_linear)))
for x, y, c in zip(xs_linear, ys_linear, colors):
    plt.scatter(x, y, color=c)
plt.show()
```

The above decision boundary may or may not be all that great for the data.  We will come back to how we can find a better way to choose the parameters of the linear decision boundary.

### Non-linear decision boundaries


```python
n = 12
x1 = 2 * np.random.rand(1,n) -1
y1 = 2 * np.random.rand(1,n) -1


r = 2
def drawcircle(x):
    a = []
    for item in x:
        a.append(math.sqrt(r**2 - item**2))
    return a

x2_a = 2 * r * np.random.rand(1,n/2) - r
y2_a = drawcircle(x2_a[0])
x2_b = 2 * r * np.random.rand(1,n/2) - r
y2_b = drawcircle(x2_b[0])
y2_b = np.multiply(y2_b,-1)

x2 = np.hstack((x2_a, x2_b))
y2 = np.hstack((y2_a, y2_b))

xs = np.array([x1, x2])
ys = np.array([y1, y2])

colors = cm.rainbow(np.linspace(0, 1, len(ys)))

for x, y, c in zip(xs, ys, colors):
    plt.scatter(x, y, color=c)
from pylab import axes
axes().set_aspect('equal', 'datalim')
plt.show()
```

Clearly with the above dataset we will want a non-linear decision boundary.

We will use some intuition and realize that since the two groups above can be separated by a circle, we'll use a circle:

$$h_\theta(x) = g(\theta_0 + \theta_1x_1 + \theta_2x_2 + \theta_3x_1^2 + \theta_4x_2^2)$$

Again we will just guess the paramteres (will later talk about how to automatically choose the parameters):

$$
\theta = 
\begin{bmatrix}
-2\cr
0\cr
0\cr

1
\end{bmatrix}$$

Where we predict $y=1$ if $-2 + x_1^2 + x_2^2 \geq 0$

### Cost function for logistic regression

If we want to do a better job of choosing our parameters for our decision boundaries, we first must have a measure of how well our decision boundaries are doing.  That is the cost function.

The cost function is a function of the dataset and parameters. These are of the form:

- Training set with m samples: $\{ (x^{(1)}, y^{(1)}), (x^{(2)}, y^{(2)}), ..., (x^{(m)}, y^{(m)})\}$
- Where each sample is a vector of the form

$$
x \in 
\begin{bmatrix}
x_0\cr
x_1\cr
...\cr
x_n
\end{bmatrix}, x_0 = 1, y \in \{0,1\}$$

$$ h_\theta (x) = \frac {1} {1 + e^{-\theta^T x}}$$

- Of course, we want to have a convex cost function
- Use:

$$J(\theta) = \frac{1}{m} \sum\limits_{i=1}^{m} \text{Cost} (h_\theta(x^{(i)}), y^{(i)})$$

$$ \text{Cost} (h_\theta(x), y) =
\begin{cases}
-\log(h_\theta(x)) & \mbox{if } y = 1 \\
-\log(1 - h_\theta(x)) & \mbox{if } y = 0
\end{cases}$$



### Gradient descent for logistic regression

First let's define simplified cost function:

$$\text{Cost} (h_\theta(x), y) =  -y \log(h_\theta(x)) -(1-y) \log(1 - h_\theta(x)) $$

(since $y=1$ or $y=0$ always, the above is a more compact way of writing the cost function)

Which gives us:

$$J(\theta) = \frac{1}{m} \Big[ \sum\limits_{i=1}^{m} -y^{(i)} \log(h_\theta(x^{(i)})) -(1-y^{(i)}) \log(1 - h_\theta(x^{(i)})) \Big] $$

To fit paramters $\theta$:

$$\underset { \theta} {min}  \ J(\theta)$$

To make a prediction given new x:

$$ \text{output} \ \ \ h_\theta (x) = \frac {1} {1 + e^{-\theta^T x}}$$

(remember that the above is equivalently:

$$ h_\theta(x) = P (y = 1 \ | \ x;  \theta) $$

#### Gradient descent algorithm

Repeat {

$\theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta) $

}

Where 

$$ \frac{\partial}{\partial \theta_j} J(\theta) = \frac{1}{m} \sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} $$

(where $\theta_j$ all updated simultaneously)

Algorithm is identical to linear regression!

```python
### Decision boundary example 2 - with gradient descent
import matplotlib.cm as cm

m = 14
J = 2

theta = np.array([-4.0, 1.0, 1.0])

x = np.zeros((J+1,m))
x[0] = 1
print x

y_0 = np.zeros((1,m/2))[0]
y_1 = np.ones((1,m/2))[0]
y = np.hstack((y_0, y_1))
print y

for i in range(m):
    x[1][i] = 2 * np.random.rand(1,1) + 1.5 * y[i]
for i in range(m):
    x[2][i] = 2.5 * np.random.rand(1,1) + 1 * y[i]
print x


linex = np.arange(0, 3.5, 0.2)
liney = (- theta[1] * linex - theta[0] * 1) / theta[2]
plt.plot(linex, liney, color='k')
for x1, x2, yi in zip(x[1], x[2], y):
    if yi: plt.scatter(x1, x2, color='r')
    else: plt.scatter(x1, x2, color='b')

plt.show()
```

### --- implement h_theta ---

First let's implement our logistic function.  We will print out h_theta(x) for each training sample, and see if it matches our expectations.

```python
def h_theta(x, theta):
    def dotproduct(x, theta):
        return np.dot(theta,x)
    return 1/(1+math.exp(-np.dot(x, theta)))

h = np.zeros((1,m))[0]
for i in range(m):
    h[i] = h_theta(x[:,i],theta)
    
print h
```

### --- implement cost function ---

Next let's implement our cost function:

$$J(\theta) = \frac{1}{m} \Big[ \sum\limits_{i=1}^{m} -y^{(i)} \log(h_\theta(x^{(i)})) -(1-y^{(i)}) \log(1 - h_\theta(x^{(i)})) \Big] $$

```python
def Cost(x, theta):
    costtotal = 0
    for i in range(m):
        samplecost = -y[i] * math.log(h_theta(x[:,i],theta)) - (1 - y[i]) * math.log(1 - h_theta(x[:,i],theta))
        costtotal = costtotal + samplecost
    return (1.0 / m) * costtotal
```

### --- implement gradient descent ---

Next let's implement our gradient descent algorithm:

Repeat {

$\theta_j := \theta_j - \alpha \frac{\partial}{\partial \theta_j} J(\theta) $

}

Where 

$$ \frac{\partial}{\partial \theta_j} J(\theta) = \frac{1}{m} \sum\limits_{i=1}^{m} (h_\theta(x^{(i)}) - y^{(i)}) x_j^{(i)} $$

(where $\theta_j$ all updated simultaneously)

Remember: very useful to plot cost as a function of iterations!  (Should be nicely decreasing)


```python
## Return to linear decision boundary problem

def GradDescent(x,theta):
    alpha = .5
    numiter = 100
    iterations = np.arange(0, numiter)
    costs = np.zeros((1,numiter))[0]
    for itr in iterations:
        #print 'itr number is '; print itr
        prev_theta = theta    
        for j in range(J+1):
            total = 0
            for i in range(m):
                #print 'h_theta is'; print h_theta(x[:,i],prev_theta)
                #print 'y is'; print y[i]
                product = (h_theta(x[:,i],prev_theta) - y[i]) * x[j][i]
                total = total + product
            #print 'theta[j] is ' + str(theta[j])
            #print 'total is ' + str(total)
            #print 'alpha is ' + str(alpha)
            theta[j] = prev_theta[j] - alpha * (1.0 / m) * total
            #print 'theta[j] now set to ' + str(theta[j])
        #print 'Cost is'; print Cost(x,theta)
        costs[itr] = Cost(x,theta)

        # Uncomment to print after each iteration
        linex = np.arange(0, 3.5, 0.2)
        liney = (- theta[1] * linex - theta[0]) / theta[2]
        plt.plot(linex, liney, color='k')
        plt.plot(linex, liney, color='k')
        for x1, x2, yi in zip(x[1], x[2], y):
            if yi: plt.scatter(x1, x2, color='r')
            else: plt.scatter(x1, x2, color='b')
        plt.axis([-0.5, 3.5, 0, 4])
        plt.show() 
    print iterations
    print costs
    plt.plot(iterations,costs)
    #plt.axis([0, 100, 0, 0.2])
    plt.show()
    
GradDescent(x,theta)

linex = np.arange(0, 3.5, 0.2)
liney = (- theta[1] * linex - theta[0]) / theta[2]
plt.plot(linex, liney, color='k')
plt.plot(linex, liney, color='k')
for x1, x2, yi in zip(x[1], x[2], y):
    if yi: plt.scatter(x1, x2, color='r')
    else: plt.scatter(x1, x2, color='b')
plt.axis([-0.5, 3.5, 0, 4])
plt.show() 
```
